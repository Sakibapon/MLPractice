```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```

```python
sns.distplot(df['tip'],kde=False,bins=20)
```

```python
sns.countplot('sex',data=df)
```

```python
sns.barplot(y='total_bill',x='smoker',data=df)
```

```python
sns.boxplot('day','total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df)
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

```python
sns.violinplot(x="total_bill", y="day", data=df)
```
