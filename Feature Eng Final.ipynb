```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('https://raw.githubusercontent.com/krishnaik06/Advanced-House-Price-Prediction-/master/train.csv')
print(dataset.shape)
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(dataset,dataset['SalePrice'],test_size=0.1,random_state=0)
X_train.shape,X_test.shape
```

```python
nafeatures=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype=='O']
print(len(nafeatures))
for feature in nafeatures:
    print(feature,'Has',np.round(dataset[feature].isnull().mean(),4),'% Missing Vale')
```

```python
def abc(dataset,nafeatures):
    data=dataset.copy()
    data[nafeatures]=data[nafeatures].fillna('Missing')
    return data
dataset=abc(dataset,nafeatures)
dataset[nafeatures].isnull().sum()
```

```python
catafeatures=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype!='O']
print(catafeatures)
for feature in catafeatures:
    print(feature,'Has',np.round(dataset[feature].isnull().mean(),4),'% Missing value')
```

```python
for feature in catafeatures:
    median=dataset[feature].median()
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median,inplace=True)
dataset[catafeatures].isnull().sum()
```

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    dataset[feature]=dataset['YrSold']-dataset[feature]
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
numfeatures=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']
for feature in numfeatures:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset[['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].head()
```

```python
categfeatures=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
len(categfeatures)
```

```python
for feature in categfeatures:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset[feature])
    temp_df=temp[temp>0.1].index
    dataset[feature]=np.where(dataset[feature].isin (temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head()
```

```python
scalingfeatures=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
print(len(scalingfeatures))
scalingfeatures
```

```python
from sklearn.preprocessing import LabelEncoder
sc=LabelEncoder()
for feature in scalingfeatures:
    dataset[feature]=sc.fit_transform(dataset[feature])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
scaler.fit(dataset[scalingfeatures])
```

```python
scaler.transform(dataset[scalingfeatures])
```

```python
data=pd.concat([dataset[['Id','SalePrice']].reset_index(drop=True),pd.DataFrame(scaler.transform(dataset[scalingfeatures]),
                                                                                columns=scalingfeatures)],axis=1)
```
