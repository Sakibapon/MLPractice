```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Lasso
```

```python
dataset=pd.read_csv('ttt.csv')
dataset=dataset.drop(['Unnamed: 0'],axis=1)
dataset.head()
```

```python
X_train=dataset.drop(['Id','SalePrice'],axis=1)
y_train=dataset[['SalePrice']]
#y_train.head()
```

```python
sel=SelectFromModel(Lasso(alpha=0.005,random_state=0))
sel.fit(X_train,y_train)
```

```python
sel.get_support()
```

```python
selectedfeatures=X_train.columns[(sel.get_support())]
print('Total Features :' ,len(X_train.columns))
print('Selected Features :',len(selectedfeatures))
```

```python
X_train=X_train[selectedfeatures]
X_train.head()
```
